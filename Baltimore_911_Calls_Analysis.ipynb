<p style="font-family: Calibri; font-size:1.3em;color:green;"> <em> <b> We need to drop columns which won't add much value to identifying trends in data.
<br> <br> CouncilDistrict and PoliceDistrict have 1:1 mapping i.e 1.0 for Southeastern police district, 2.0 for Northeastern police district etc. So we can omit CouncilDistrict.
<br> <br> RecordID, CallNumber, ZipCode, Census_Tracts don't look like they have much of an impact.
<br> <br> Location column is too detailed, and district info should be enough for us to identify trends region-wise.
<br> <br> IncidentLocation column is again too detailed, and neighbourhood along with district info should be enough for us to identify trends region-wise.
<br> <br> SheriffDistricts too can be omitted doesn't really tell us anything in particular.
<br> <br> Neighbourhood and Community_Statistical_Areas are similar, so we can remove the Community_Statistical_Areas' column
<br> <br> For VRIZones, most of the values are NaN. Makes sense to omit this column as well. </b> </em> </p>

```python
df.drop('RecordID', axis=1, inplace=True)
df.drop('CallNumber', axis=1, inplace=True)
df.drop('ZipCode', axis=1, inplace=True)
df.drop('Census_Tracts', axis=1, inplace=True)
df.drop('VRIZones', axis=1, inplace=True)
df.drop('Location', axis=1, inplace=True)
df.drop('CouncilDistrict', axis=1, inplace=True)
df.drop('IncidentLocation', axis=1, inplace=True)
df.drop('PolicePost', axis=1, inplace=True)
df.drop('SheriffDistricts', axis=1, inplace=True)
df.drop('Community_Statistical_Areas', axis=1, inplace=True)
df
```

<p style="font-family: Calibri; font-size:1.3em;color:green;"> <em> <b> Good time to remove outliers i.e incidents which occurred less than a 1000 times. In a dataset of 66 million records, removing incidents less than 1000 occurrences won't matter. </b> </em> </p>

```python
df.Description.value_counts()
```

```python
value_counts = df['Description'].value_counts()
to_remove = value_counts[value_counts <= 1000].index
df = df[~df.Description.isin(to_remove)]
df.shape
```

```python
df['Description'].value_counts()
```

<p style="font-family: Calibri; font-size:1.3em;color:green;"> <em> <b> So we removed about 147k records. Now checking how the description data looks like...</b> </em> </p>

```python
df.Description.unique()
```

<p style="font-family: Calibri; font-size:1.3em;color:green;"> <em> <b> For filtering data based on the 'Description'...
<br> <br> We can remove all rows containing 'Unknown' in the 'Description' column.
<br> <br> We can also remove 'Other' descriptions.
<br> <br> Remove descriptions like '5G', '7C', '56'.
<br> <br> Remove 911/HANGUP, 911/No Voice and MISTAKE.
<br> <br> Moreover, we can combine all Larceny related incidents to have a common 'Larceny' description.
<br> <br> </b> </em> </p>

```python
df = df[df['Description'] != 'Other']
df = df[df['Description'] != '7C']
df = df[df['Description'] != '5G']
df = df[df['Description'] != '56']
df = df[df['Description'] != '911/HANGUP']
df = df[df['Description'] != '911/No Voice']
df = df[df['Description'] != 'MISTAKE']
df = df[df['Description'] != '911/NO']
df = df[df['Description'] != '*INV']
df = df[df['Description'] != '*POSSIBLE']
df = df[df['Description'] != '*INVESTIGATE']
df = df[df['Description'] != '*INVEST']
df.shape
```

```python
df = df[~df.Description.str.contains("UNKNOWN")]
df.shape
```

```python
df.loc[df['Description'].str.contains('larceny', case=False), 'Description'] = 'Larceny'
df.shape
```

<p style="font-family: Calibri; font-size:1.3em;color:green;"> <em> <b> For scatter plots, the columns need to have numeric values.
<br> Mapping priority and PoliceDistrict to numerical values and adding those extra columns in the dataframe </b> </em> </p>

```python
print(df.Priority.unique())
print(df.PoliceDistrict.unique())
df["NumericPriority"] = df.Priority
df["PoliceDistrictNumeric"] = df.PoliceDistrict
df['NumericPriority'] = df['NumericPriority'].map({'High': 1, 'Medium': 2, 'Low': 3, 'Non-Emergency': 4})
df['PoliceDistrictNumeric'] = df['PoliceDistrictNumeric'].replace(['Northeastern', 'Northern', 'Southwestern', 'Southern', 'Central', 'Southeastern', 'Eastern', 'Western', 'Northwestern'], [0,1,2,3,4,5,6,7,8])
df
```

<p style="font-family: Calibri; font-size:1.3em;color:green;"> <em> <b> Trying to find a relation between Priority of the call, PoliceDistrict where it was received and Call time.

<br> So, our predicted variable (dependent variable y) should be Priority,
<br> our predictor variables (independent variables x1, x2) should be CallTime and PoliceDistrict.
</b> </em> </p>

```python
plt.rcParams["figure.figsize"] = 25,25
plt.rc('xtick',labelsize=25)
plt.rc('ytick',labelsize=25)
# df.plot(ms=3)
df.plot.scatter('CallTime', 'NumericPriority', c='PoliceDistrictNumeric', colormap='viridis', s=400, sharex=False)
```
